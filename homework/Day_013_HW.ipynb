### 3. [簡答題] 承上題，請問為什麼依照 index 合併會發生錯誤？請用程式解決。

````python
因為原本的欄位重複，導致無法正確合併。
````

````python
df1 = pd.DataFrame({
    'fruit_df1': ['apple', 'banana', 'orange'] * 3,
    'weight': ['high', 'medium', 'low'] * 3,
    'price': np.random.randint(0, 15, 9)
})

df2 = pd.DataFrame({
    'fruit_df2': ['apple', 'orange', 'pine'] * 2,
    'weight': ['high', 'low'] * 3,
    'price': np.random.randint(0, 15, 6)
})
print(df1)
print('-'*40)
print(df2)

pd.merge(df1, df2, left_on='fruit_df1', right_on='fruit_df2')
````
