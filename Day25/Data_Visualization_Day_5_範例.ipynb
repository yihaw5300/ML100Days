# 本地運行範例程式由此開始

```python
### 利用 PANDAS 取得酒的品質資料
# 請記得把CSV檔放在當前的工作目錄 

#如何察知工作目錄:
import os
os.getcwd()

df_red = pd.read_csv("winequality_red.csv")
df_white = pd.read_csv("winequality_white.csv")
```

```python
#資料整理
df_red["color"] = "R"
df_white["color"] = "W"

#整合紅酒與白酒的資料
df_all=pd.concat([df_red,df_white],axis=0)

# 檢查合併後的資料集
df_all.head()
```

```python
# 重新命名特徵
df_all.rename(columns={'fixed acidity': 'fixed_acidity','citric acid':'citric_acid',
                       'volatile acidity':'volatile_acidity','residual sugar':'residual_sugar',
                       'free sulfur dioxide':'free_sulfur_dioxide',
                       'total sulfur dioxide':'total_sulfur_dioxide'}, inplace=True)
# 檢查合併後的資料集
df_all.head()
```

```python
# 資料整理之前
#酒類的品質是首要, 所以我們可以以 quality 為 Y label
sns.histplot(df_all["quality"])
```

```python
#處理缺失值
df = pd.get_dummies(df_all, columns=["color"])
df_all.isnull().sum()
```

```python
#可以使用 info() 方法瞭解有關資料集屬性的更多資訊。特別是行和列的數量、列名稱、它們的數據類型和空值數。
df_all.info()
```

```python
#要瞭解數據集的統計摘要,即記錄數、平均值、標準差、最小值和最大值,我們使用描述()。
df_all.describe()
```

```python
#所有數值數據數值不一, 難以由圖表一眼看清
#在垂直軸上計數,在水平軸上使用值範圍。hist 函數通過將所有屬性繪製在一起使操作變得簡單。
df_all.hist(bins=10, color='lightblue',edgecolor='blue',linewidth=1.0,
          xlabelsize=8, ylabelsize=8, grid=False)    
```

# tight_layout 介紹

上圖, 我們發現 label 被重疊到了, 若要逐一調整十分麻煩,

這時, 可以考慮tight_layout

tight_layout提供rect參數，表示一個外界的框框:(x1, y1, x2, y2)默認是(0, 0, 1, 1)

(x1, y1)矩形限制框左下角座標

(x2, y2)矩形限制框右上角座標

```python
#建立一個Figure（空的顯示區）
fig = plt.figure()

#直接製圖在剛剛建立的figure
ax1 = plt.subplot(221)
ax2 = plt.subplot(223)
ax3 = plt.subplot(122)

#避免多個圖重疊，使用tight_layout分開, 可以節省新增Figure的軸的動作
plt.tight_layout()
```

```python
df_all.hist(bins=10, color='lightblue',edgecolor='blue',linewidth=1.0,
          xlabelsize=8, ylabelsize=8, grid=False)    

plt.tight_layout(rect=(0, 0, 1.2, 1.2))
```

# 熱力圖

是一個以顏色變化來顯示數據的矩陣。簡單來說，就是用依據數字的不同，使用不同的顏色來呈現數據

利用熱力圖可以看資料表裡多個特徵兩兩的相似度。


```python
# 最直接的做法是把資料集丟給heatmap函數
sns.heatmap(df_all.corr())

# 可以看到熱力圖主要展示的是二維數據的數據關係
# 不同大小的值對應不同的顏色深淺
# df_all.corr() - 把 string transfer to float type 
```

```python
# 加上標記data
sns.heatmap(df_all.corr(), annot=True)
```

```python
# 這個資料集包括了平均數, 標準差, 一定區間的數值
# 所以, 我們可以考慮使用 bootstrap 幫忙繪製圖表包含 mean, median and mid-range statistics.
s = pd.Series(df_all['quality'])
pd.plotting.bootstrap_plot(s, samples=500)
plt.tight_layout(rect=(0, 0, 1.2, 1.2))
```
